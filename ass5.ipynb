```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=load_iris()
data
target=data.target
df=data.data
```

```python
df=pd.DataFrame(data=df[:, :], columns=data.feature_names[:])
```

```python
y=data.target
```

```python
df
```

```python
df=df.drop(["sepal width (cm)", "sepal length (cm)"], axis=1)
```

```python
c=data.target
plt.scatter(df["petal length (cm)"], df["petal width (cm)"],c=c)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2)
```

```python
from sklearn import svm
clf=svm.SVC(kernel="linear", C=1)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(np.array(X_test), np.array(y_test), clf=clf, legend=1)
```

```python
clf=svm.SVC(kernel="rbf", gamma=0.7, C=1)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_decision_regions(np.array(X_test), np.array(y_test), clf=clf, legend=1)
```

```python
clf=svm.SVC(kernel="poly", degree=3, C=1)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_decision_regions(np.array(X_test), np.array(y_test), clf=clf, legend=1)
```
